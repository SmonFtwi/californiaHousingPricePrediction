# **1.   Import dependency**

In the correlation dataset, certain features shows a weak correlation with the target dataset, such as 'AveBedrms,' 'AveRooms,' and 'AveOccup.' To enhance the features, we will engage in feature engineering by manipulating existing features. In the upcoming code, we will introduce the 'bedroom_ratio,' calculated as ['AveBedrms'] divided by df['AveRooms'], and 'household_rooms,' calculated as ['AveRooms'] divided by df['AveOccup'].

```python
#feature engineering
df['Bedroom_ratio'] = df['AveBedrms'] / df['AveRooms']
df['household_rooms']= df['AveRooms'] / df['AveOccup']
```

```python
correlation_matrix = df.corr()
# Plotting the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, cmap='coolwarm', annot=True, fmt=".2f", linewidths=.5)
plt.title('Correlation Heatmap')
plt.show()
```




# **3. Data preprocessing**



```python
df['AveRooms'] = np.log(df['AveRooms']+1)
df['Price'] = np.log(df['Price']+1)
df['Latitude'] = np.log(df['Latitude']+1)
df['AveBedrms']= np.log(df['AveBedrms']+1)
```

```python
x = df.drop(['Price'], axis=1)
y=df['Price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
scalar= StandardScaler()
x_train= scalar.fit_transform(x_train)
x_test= scalar.transform(x_test)
```

# **4. Model Training**

```python
rf= RandomForestRegressor(n_estimators=100, random_state=22,max_depth=12)
rf.fit(x_train,y_train)
```

```python
rf.score(x_train,y_train)
```

```python
rf.score(x_test,y_test)
```

# **5. Model Evaluation**

```python
y_pred= rf.predict(x_test)
test_mseRf= mean_squared_error(y_test,y_pred)
test_mseRf
```

# **5. Model Saving**

```python

pickle.dump(rf, open('rgmodel.pkl', 'wb'))
```
